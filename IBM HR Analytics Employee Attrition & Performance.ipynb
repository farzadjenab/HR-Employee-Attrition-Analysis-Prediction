```python
# Convert target variable to numbers
# Yes = 1 (left the company)
# No = 0 (stayed)

df['Attrition_Numeric'] = df['Attrition'].map({'Yes': 1, 'No': 0})

print("Attrition converted to numbers:")
print(df[['Attrition', 'Attrition_Numeric']].head(10))
```

```python
# We'll select some important features based on our analysis
# Keeping it simple for a junior analyst

# First, let's see which columns are categorical and numerical
print("Categorical columns:")
print(df.select_dtypes(include=['object']).columns.tolist())

print("\nNumerical columns:")
print(df.select_dtypes(include=[np.number]).columns.tolist())
```

```python
# Select features for our simple model
# Based on our analysis, these seem important:

selected_features = [
    'Age',
    'MonthlyIncome',
    'OverTime',
    'YearsAtCompany',
    'WorkLifeBalance',
    'EnvironmentSatisfaction',
    'JobSatisfaction',
    'DistanceFromHome',
    'NumCompaniesWorked',
    'TotalWorkingYears'
]

print("Selected features for our model:")
for i, feat in enumerate(selected_features, 1):
    print(f"{i}. {feat}")
```

```python
# Create a copy of dataframe with selected features
df_model = df[selected_features + ['Attrition_Numeric']].copy()

# Convert OverTime to numbers (Yes=1, No=0)
df_model['OverTime'] = df_model['OverTime'].map({'Yes': 1, 'No': 0})

print("Data after conversion:")
df_model.head()
```

```python
# Check for any missing values
print("Missing values check:")
print(df_model.isnull().sum())

print("\nData shape:", df_model.shape)
print("\nData types:")
print(df_model.dtypes)
```

```python
# Import necessary libraries for modeling
from sklearn.model_selection import train_test_split

# Separate features (X) and target (y)
X = df_model.drop('Attrition_Numeric', axis=1)
y = df_model['Attrition_Numeric']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# Split data into training and testing sets
# 80% for training, 20% for testing

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,  # 20% for testing
    random_state=42  # For reproducibility
)

print(f"Training set size: {X_train.shape[0]} rows")
print(f"Testing set size: {X_test.shape[0]} rows")
```

```python
# Scale the features so they have similar ranges
# This helps the model perform better

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit on training data and transform both
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Data has been scaled! âœ…")
print(f"X_train_scaled shape: {X_train_scaled.shape}")
```

```python
# Import and create the model
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model
model = LogisticRegression(random_state=42, max_iter=1000)

# Train the model on training data
model.fit(X_train_scaled, y_train)

print("Model trained successfully! âœ…")
```

```python
# Make predictions on test data
y_pred = model.predict(X_test_scaled)

# Show first 20 predictions vs actual
comparison = pd.DataFrame({
    'Actual': y_test.values[:20],
    'Predicted': y_pred[:20]
})
comparison['Match'] = comparison['Actual'] == comparison['Predicted']

print("First 20 Predictions vs Actual:")
print(comparison)
```

```python
# Calculate accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
# Detailed classification report
from sklearn.metrics import classification_report

print("Classification Report:")
print("=" * 60)
print(classification_report(y_test, y_pred, target_names=['Stayed (No)', 'Left (Yes)']))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

# Visualize confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Stayed (No)', 'Left (Yes)'],
            yticklabels=['Stayed (No)', 'Left (Yes)'])
plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()

# Explain the confusion matrix
print("\nConfusion Matrix Explanation:")
print(f"âœ… True Negatives (Correctly predicted Stayed): {cm[0][0]}")
print(f"âŒ False Positives (Predicted Left but Stayed): {cm[0][1]}")
print(f"âŒ False Negatives (Predicted Stayed but Left): {cm[1][0]}")
print(f"âœ… True Positives (Correctly predicted Left): {cm[1][1]}")
```

```python
# See which features are most important for prediction
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
})

# Sort by absolute value of coefficient
feature_importance['Abs_Coefficient'] = abs(feature_importance['Coefficient'])
feature_importance = feature_importance.sort_values('Abs_Coefficient', ascending=False)

print("Feature Importance (sorted by importance):")
print("=" * 50)
print(feature_importance[['Feature', 'Coefficient']])
```

```python
# Visualize feature importance
plt.figure(figsize=(10, 6))

colors = ['#e74c3c' if x > 0 else '#2ecc71' for x in feature_importance['Coefficient']]

plt.barh(feature_importance['Feature'], feature_importance['Coefficient'], color=colors)
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.title('Feature Importance in Predicting Attrition', fontsize=14, fontweight='bold')
plt.axvline(x=0, color='black', linestyle='-', linewidth=0.5)

plt.tight_layout()
plt.show()

print("\nðŸ’¡ Interpretation:")
print("ðŸ”´ Red bars (positive): Higher values increase chance of leaving")
print("ðŸŸ¢ Green bars (negative): Higher values decrease chance of leaving")
```

```python
# Scale the new employee data
new_emp_scaled = scaler.transform(new_emp_df)

# Make prediction
prediction = model.predict(new_emp_scaled)
prediction_proba = model.predict_proba(new_emp_scaled)

print("\n" + "=" * 50)
print("PREDICTION RESULT:")
print("=" * 50)

if prediction[0] == 1:
    print("âš ï¸  This employee is likely to LEAVE")
else:
    print("âœ… This employee is likely to STAY")

print(f"\nProbability of Staying: {prediction_proba[0][0] * 100:.1f}%")
print(f"Probability of Leaving: {prediction_proba[0][1] * 100:.1f}%")
```
